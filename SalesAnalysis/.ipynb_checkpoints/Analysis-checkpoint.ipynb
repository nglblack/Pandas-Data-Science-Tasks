```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

### Merging 12 months of slaes data into a single file

```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")

files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index=False)
```

### Clean Data

```python
all_months_data = all_months_data.dropna()
all_months_data = all_months_data[all_months_data['Order Date'].str[0:2].str.isnumeric()]

all_months_data['Month'] = all_months_data['Order Date'].str[0:2].astype('int32')
all_months_data = all_months_data.reset_index(drop=True)
all_months_data.to_csv("cleaned_all_data.csv", index=False)
```

### Read the cleaned data

```python
all_data = pd.read_csv("cleaned_all_data.csv")
all_data.head()
```

### Add a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Add a city and state column

```python
all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1].strip())
all_data['State'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[2].split(' ')[1].strip())
all_data['City & State'] = all_data['City'] + ' ' + all_data['State']

all_data.head()
```

```python
results = all_data.groupby('Month').sum()
```

```python
product_quantity = all_data.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False)
```

```python
# Plotting a bar chart
plt.figure(figsize=(12, 6))
product_quantity.plot(kind='bar', color='skyblue')
plt.title('Quantity Ordered for Each Product')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

### What was the best month for sales? How much was earned that month? 

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Assuming 'all_data' is your DataFrame
monthly_sales = all_data.groupby('Month')['Sales'].sum()

# Plotting a bar chart
plt.figure(figsize=(10, 6))

# Define a function to format y-axis labels in millions
def millions_formatter(x, pos):
    return f"${x / 1e6:.2f}M"

# Apply the millions_formatter to the y-axis
formatter = FuncFormatter(millions_formatter)
plt.gca().yaxis.set_major_formatter(formatter)

monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales (Millions USD)')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.tight_layout()

# Show the plot
plt.show()

# Find the best month for sales and the corresponding amount
best_month = monthly_sales.idxmax()
sales_amount = monthly_sales.max()

print(f"The best month for sales was Month {best_month} with total sales of ${sales_amount:.2f} million.")
```

### What city had the highest number of sales

```python
city_sales = all_data.groupby('City & State')['Sales'].sum().sort_values(ascending=False)

# Plotting a bar chart
plt.figure(figsize=(12, 6))
city_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by City & State')
plt.xlabel('City & State')
plt.ylabel('Sales (USD)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()

# Identify the city with the highest sales
highest_sales_city_state = city_sales.idxmax()
highest_sales_amount = city_sales.max()

print(f"The city with the highest sales is {highest_sales_city_state} with total sales of ${highest_sales_amount:.2f}.")
```

### What products are most often sold together?

```python
all_data.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
from itertools import combinations
from collections import Counter

```

```python
count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```

```python
top_combinations = count.most_common(10)

# Extracting combination labels and counts
combination_labels, counts = zip(*top_combinations)

# Plotting the horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(range(len(combination_labels)), counts, color='skyblue')
plt.yticks(range(len(combination_labels)), combination_labels)
plt.xlabel('Count')
plt.ylabel('Combinations')
plt.title('Top 10 Combinations Sold Together')
plt.show()
```
